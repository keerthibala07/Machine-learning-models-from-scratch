```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2.1,3.8,6.5,9.2,11.3])
```

```python
x_b=np.c_[np.ones((x.shape[0],1)),x]
x_b
```

```python
alpha_ridge=1.0
alpha_lasso=0.1
num_iteration=1000
learning_rate=0.01
```

```python
def lasso_regression(x,y,alpha,num_iteration,learning_rate):
  m,n=x.shape
  theta=np.zeros(n)
  for _ in range(num_iteration):
    y_pred=x.dot(theta)
    gradient=(x.T.dot(y_pred-y)+alpha*np.sign(theta))/m
    theta-=learning_rate*gradient
  return theta
```

```python
def ridge_regression(x,y,alpha,num_iteration,learning_rate):
  m,n=x.shape
  theta=np.zeros(n)
  for _ in range(num_iteration):
    y_pred=x.dot(theta)
    gradient=(x.T.dot(y_pred-y)+alpha*theta)/m
    theta-=learning_rate*gradient
  return theta
```

```python
theta_ridge=ridge_regression(x_b,y,alpha_ridge,num_iteration,learning_rate)
theta_lasso=lasso_regression(x_b,y,alpha_lasso,num_iteration,learning_rate)
```

```python
plt.scatter(x,y,label='data')
plt.plot(x,x_b.dot(theta_ridge),label=f'Rigde (L2) Regression',color='red')
plt.plot(x,x_b.dot(theta_lasso),label=f'Lasso (L1) Regression',color='green')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
