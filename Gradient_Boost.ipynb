```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv')
test=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv')
```

```python
train.info()
test.info()
```

```python
train.set_index("PassengerId",inplace=True)
test.set_index('PassengerId',inplace=True)
```

```python
test
```

```python
y_train=train['Survived']
```

```python
x_train=train.drop('Survived',axis=1,inplace=True)
```

```python
train.shape
```

```python
train_test=train.append(test)#891 data not enough
train_test.shape
```

```python
train.isnull().sum()
```

```python
train_test
```

```python
drop=['Name','Age',"SibSp",'Ticket','Cabin','Parch','Embarked']
train_test.drop(labels=drop,axis=1,inplace=True)
```

```python
train_test
```

```python
train_tes_d=pd.get_dummies(train_test,columns=['Sex'])
```

```python
train_tes_d.shape
```

```python
train_tes_d.isnull().sum()
```

```python
train_tes_d.fillna(value=0.0,inplace=True)
```

```python
x1_train=train_tes_d.values[0:891]
x1_test=train_tes_d.values[891:]

```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train_scale=scaler.fit_transform(x1_train)
x_test_scale=scaler.transform(x1_test)#already fitted no need to give fit
```

```python
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1=train_test_split(x_train_scale,y_train,random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate = learning_rate, max_features=2, max_depth = 2, random_state = 0)
    gb.fit(x_train1, y_train1)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb.score(x_train1, y_train1)))
    print("Accuracy score (testing): {0:.3f}".format(gb.score(x_test1, y_test1)))
    print()
```
